```python
import torch
from diffusers import DiffusionPipeline

# Switch to "mps" for apple devices
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", dtype=torch.float16, device_map="cuda")

# Memory optims
pipe.enable_attention_slicing("max")
pipe.enable_vae_slicing()
pipe.enable_vae_tiling()

# Set desired prompt and generate image
prompt = "Top-down view image of an abandoned tower for a video game"
image = pipe(prompt, height=480, width=480, num_inference_steps=25).images[0]
```
